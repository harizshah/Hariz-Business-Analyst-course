```python
#Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import scipy.stats as ss
from causalinference import CausalModel
```

```python
#Loading Data
df = pd.read_csv("stackoverflow.csv")
df.head()
```

#Data Analysis

```python
#Picking variables
df = df.iloc[:,:8]
df.head(0)
```

```python
#Transforming character variables
df = pd.get_dummies(df, drop_first = True)
df.head(1)
```

```python
#Looking at Group Means
df.groupby("Remote_Remote").mean()
```

```python
#T-tests
#stating with variables to test
continuous = ['Salary', 'YearsCodedJob']

#Where to store results
stat = {}
pvalue = {}

#Loop
for x in continuous:
  group1 = df.where(df.Remote_Remote == 0).dropna()[x]
  group2 = df.where(df.Remote_Remote == 1).dropna()[x]
  stat[x], pvalue[x] = ss.ttest_ind(group1, group2)
ttests = pd.DataFrame.from_dict(pvalue, orient = 'Index')
ttests.columns = ['pvalue']
print(ttests)
```

#Matching

```python
#Isolate y, treat, confounders
y = df.CareerSatisfaction.values
treat = df.Remote_Remote.values
confounders = df.drop(columns = ["Remote_Remote",
                                 "CareerSatisfaction"]).values
```

```python
#Matching
model = CausalModel(y, treat, confounders)
model.est_via_matching(bias_adj = True)
print(model.estimates)
```

ATE (Average Treatment Effect): The estimated treatment effect size for the entire treated group is approximately 0.114, but the standard error is quite high (0.130). The 'z' score (0.881) suggests that this effect is not statistically significant (p-value = 0.378). The 95% confidence interval ranges from -0.140 to 0.368, indicating a wide range of potential effects.

ATC (Average Treatment Effect on the Control Group): This represents the effect of the treatment on the control group. The estimated effect size is 0.106, but like the ATE, it's not statistically significant (z = 0.778, p-value = 0.436). The confidence interval spans from -0.161 to 0.372.

ATT (Average Treatment Effect on the Treated): This shows the effect of the treatment specifically on the group that received the treatment. The estimated effect size is 0.187, but again, it's not statistically significant (z = 1.321, p-value = 0.187). The confidence interval ranges from -0.090 to 0.464.

#Robustness Check

```python
#Remove 1 confounder
confounders = df.drop(columns = ["Remote_Remote",
                                 "CareerSatisfaction",
                                 "Hobby"]).values
```

```python
#Matching
model = CausalModel(y, treat, confounders)
model.est_via_matching(bias_adj = True)
print(model.estimates)
```

ATE (Average Treatment Effect): The estimated treatment effect size for the entire treated group is approximately 0.148. The standard error is 0.131, indicating some uncertainty. The 'z' score (1.128) suggests that this effect is not statistically significant (p-value = 0.259). The 95% confidence interval ranges from -0.109 to 0.406.

ATC (Average Treatment Effect on the Control Group): This represents the effect of the treatment on the control group. The estimated effect size is 0.140, with a standard error of 0.138. The 'z' score (1.013) suggests that this effect is also not statistically significant (p-value = 0.311). The confidence interval spans from -0.131 to 0.411.

ATT (Average Treatment Effect on the Treated): This shows the effect of the treatment specifically on the group that received the treatment. The estimated effect size is 0.220, with a standard error of 0.137. The 'z' score (1.602) suggests that this effect is not statistically significant either (p-value = 0.109). The confidence interval ranges from -0.049 to 0.488.

In this robustness check, it appears that the treatment effects are not statistically significant for any of the groups (ATE, ATC, or ATT). This means that the observed effects may not be reliable or may not be due to the treatment itself. Researchers use robustness checks to see if their results hold up under different conditions, and in this case, the effects remain inconclusive in terms of statistical significance.
